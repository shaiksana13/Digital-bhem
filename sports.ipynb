```python
# 🎨 Step 1: Install Required Libraries
!pip install pandas matplotlib seaborn numpy --quiet

# 🧠 Step 2: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setup styling
sns.set_style("darkgrid")
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 13
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams["figure.facecolor"] = "#f0f2f5"

# 📂 Step 3: Load CSV Files
train_df = pd.read_csv('/content/deliveries.csv')
test_df = pd.read_csv('/content/matches.csv')

# 🔗 Step 4: Combine Datasets
df = pd.concat([train_df, test_df], ignore_index=True)

# 🔍 Step 5: Basic Info
print("📊 Data Preview:")
display(df.head())

print(f"\n🧾 Total Matches: {len(df)}")
print(f"🔎 Columns: {list(df.columns)}")

# 🧹 Step 6: Clean Missing Data
missing = df.isnull().sum()
print("\n🚨 Missing Values:\n", missing[missing > 0])

if missing.sum() > 0:
    df.dropna(inplace=True)
    print(f"✅ Cleaned missing data. Remaining rows: {df.shape[0]}")

# 🏆 Step 7: Most Successful Teams
if 'winner' in df.columns:
    team_wins = df['winner'].value_counts()

    plt.figure(figsize=(12, 6))
    sns.barplot(x=team_wins.index, y=team_wins.values, palette='cubehelix', edgecolor='black')
    plt.title("🏆 Most Successful IPL Teams", fontsize=18, weight='bold')
    plt.xlabel("Team")
    plt.ylabel("Wins")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 🌟 Step 8: Top Performing Players
if 'player_of_match' in df.columns:
    top_players = df['player_of_match'].value_counts().head(10)

    plt.figure(figsize=(10, 5))
    sns.barplot(x=top_players.values, y=top_players.index, palette='coolwarm', edgecolor='black')
    plt.title("🌟 Top 10 Players (Player of the Match Awards)", fontsize=16)
    plt.xlabel("Awards")
    plt.ylabel("Player")
    plt.tight_layout()
    plt.show()

# 🎯 Step 9: Toss Impact Analysis
if 'toss_winner' in df.columns and 'winner' in df.columns and len(df) > 0:
    toss_win = df[df['toss_winner'] == df['winner']]
    toss_win_percent = (len(toss_win) / len(df)) * 100
    print(f"\n🎯 Toss Winner also won the match in: 🌟 {toss_win_percent:.2f}% of games.")
else:
    print("⚠️ Toss winner or match winner data missing.")

# 📈 Step 10: Correlation Matrix
numeric_cols = df.select_dtypes(include=np.number).columns
if len(numeric_cols) > 1:
    plt.figure(figsize=(10, 7))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
    plt.title("🔗 Correlation Matrix", fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ Not enough numeric columns for correlation heatmap.")

# 💡 Step 11: Endorsement Recommendations
if 'player_of_match' in df.columns:
    top_endorse = df['player_of_match'].value_counts().head(5)
    print("\n📢 Recommended Players for Endorsements:")
    for i, (player, count) in enumerate(top_endorse.items(), 1):
        print(f"#{i} ⭐ {player}: {count} Player of Match awards")
else:
    print("⚠️ Cannot compute endorsements. 'player_of_match' column missing.")

# 💾 Step 12: Save Cleaned Data (Optional)
df.to_csv('/content/ipl_cleaned_data.csv', index=False)
print("\n✅ Cleaned data saved to 'ipl_cleaned_data.csv'")
```
