```python
# üé® Step 1: Install Required Libraries
!pip install pandas matplotlib seaborn numpy --quiet

# üß† Step 2: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setup styling
sns.set_style("darkgrid")
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 13
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams["figure.facecolor"] = "#f0f2f5"

# üìÇ Step 3: Load CSV Files
deliveries_df = pd.read_csv('/content/deliveries.csv')
matches_df = pd.read_csv('/content/matches.csv')

# üîó Step 4: Merge Datasets
df = deliveries_df.merge(matches_df, how='left', left_on='match_id', right_on='id')

# üîç Step 5: Basic Info
print("üìä Data Preview:")
display(df.head())

print(f"\nüßæ Total Deliveries: {len(df)}")
print(f"üîé Columns: {list(df.columns)}")

# üßπ Step 6: Clean Missing Data
missing = df.isnull().sum()
print("\nüö® Missing Values:\n", missing[missing > 0])

# Drop rows with missing data in critical match-level columns only
critical_cols = ['winner', 'player_of_match', 'toss_winner']
df.dropna(subset=critical_cols, inplace=True)
print(f"‚úÖ Cleaned critical missing data. Remaining rows: {df.shape[0]}")

# üèÜ Step 7: Most Successful Teams
team_wins = df['winner'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.index, y=team_wins.values, palette='cubehelix', edgecolor='black')
plt.title("üèÜ Most Successful IPL Teams")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# üåü Step 8: Top Performing Players
top_players = df['player_of_match'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_players.values, y=top_players.index, palette='coolwarm', edgecolor='black')
plt.title("üåü Top 10 Players (Player of the Match Awards)")
plt.xlabel("Awards")
plt.ylabel("Player")
plt.tight_layout()
plt.show()

# üéØ Step 9: Toss Impact Analysis
toss_win = df[df['toss_winner'] == df['winner']]
toss_win_percent = (len(toss_win) / len(df)) * 100
print(f"\nüéØ Toss Winner also won the match in: üåü {toss_win_percent:.2f}% of games.")

# üìà Step 10: Correlation Matrix
numeric_cols = df.select_dtypes(include=np.number).columns
if len(numeric_cols) > 1:
    plt.figure(figsize=(10, 7))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
    plt.title("üîó Correlation Matrix")
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Not enough numeric columns for correlation heatmap.")

# üí° Step 11: Endorsement Recommendations
top_endorse = df['player_of_match'].value_counts().head(5)
print("\nüì¢ Recommended Players for Endorsements:")
for i, (player, count) in enumerate(top_endorse.items(), 1):
    print(f"#{i} ‚≠ê {player}: {count} Player of Match awards")

# üíæ Step 12: Save Cleaned Data (Optional)
df.to_csv('/content/ipl_cleaned_data.csv', index=False)
print("\n‚úÖ Cleaned data saved to 'ipl_cleaned_data.csv'")
```
